Tuple is a immutable data structure, it means we cannot modify tuple directly. So, if the need is to modify the tuple we can firstly convert the tuple to a list and then make the necessary changes and finally we can revert it back to tuple.

```python
lst1 = list(tuple1)
lst1
```

```python
lst1.append(45)
lst1
```

```python
tuple1 = tuple(lst1)
tuple1
```

Set Operations

Union Operation in Set

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

union_set = set1 | set2
union_set
```

```python
union_set = set1.union(set2)
print(union_set)
```

Intersection in Set

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

intersection_set = set1 & set2
intersection_set
```

Difference in Set

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}


## return all those elements which are available in the set 1 but not in the set 2
difference_set = set1 - set2
print(difference_set)
```

```python
difference_set = set2 - set1
print(difference_set)
```

Symmetric Difference in Set

```python
set1 = {1, 2, 3}
set2 = {3, 4, 5}

set1.symmetric_difference(set2)
```

```python
symmetric_difference_set = set1 ^ set2
print(symmetric_difference_set)
```

Shallow Copy vs Deep Copy

```python
import copy

original = [[1, 2], [3, 4]]

shallow = copy.copy(original)
deep = copy.deepcopy(original)

original[0][0] = 99
print(shallow)
```
